```
sequences = [seq1] + [seq1] + [seq3] + [seq4] + [seq5] + [seq6] + [seq7] + [seq8] + [seq9] + [seq10]
print(len(sequences))
```

```
plot_options = dict(width=450,
                    plot_height=250,
                    tools='pan,wheel_zoom,reset,save')

test_plot = figure(**plot_options)
test_plot.line(range(len(seq1)),
                  seq1,
                  line_width=2)
  
test_plot.xaxis.axis_label = 'value'
test_plot.yaxis.axis_label = 'index'

show(test_plot)
```

```
string_len = len(sequences[0])

def display_str(string, i):
    return ('String ' + str(i) +' "' + ''.join([str(d) for d in string]) + '"')

min_b_ent = []

for idx,seq in enumerate(sequences):
    ks = range(1, string_len)
    entropies = []
    for k in ks:
        entropies.append(block_entropy(seq, k=k))
    
    min_b_ent.append(np.min(entropies))

    plot = figure(**plot_options)
    plot.line(ks, entropies)
    plot.title.text = display_str(seq, idx)
    plot.xaxis.axis_label = "Block size"
    plot.yaxis.axis_label = "Block entropy"
    show(plot)
```

```
# Miniminum Entropy
plot = figure(**plot_options)
plot.line(range(1,len(sequences)+1), min_b_ent)
plot.title.text = "Minimum Entropy per String"
plot.xaxis.axis_label = "String Index"
plot.yaxis.axis_label = "Minimum Entropy"
show(plot)
```

```
# Compression Schemes Analysis
c = Compressor()

gzip_length = []
bz2_length = []
lzma_length = []

for seq in sequences:
  strbyte = ''.join([str(s) for s in seq]).encode('utf-8')
  c.use_gzip()
  gzip_length.append(len(c.compress(strbyte)))

  c.use_bz2()
  bz2_length.append(len(c.compress(strbyte)))

  c.use_lzma()
  lzma_length.append(len(c.compress(strbyte)))


plot = figure(**plot_options)
plot.line(range(1,len(sequences)+1), gzip_length, color='red', legend_label='gzip')
plot.line(range(1,len(sequences)+1), bz2_length, color='green', legend_label='bz2')
plot.line(range(1,len(sequences)+1), lzma_length, color='blue', legend_label='lzma')

plot.legend.location = "top_left"

plot.title.text = "Compression Length Analysis"
plot.xaxis.axis_label = "String Index"
plot.yaxis.axis_label = "Compression Length"
show(plot)
```

```
# CTM Analysis
bdm = BDM(ndim=12,ctmname='CTM-B2-D12')
bdm_vals = []

for seq in sequences:
  bdm_vals.append(bdm.bdm(np.array(seq, dtype=int)))

plot = figure(**plot_options)
plot.line(range(1,len(sequences)+1), bdm_vals)
plot.title.text = "CTM"
plot.xaxis.axis_label = "String Index"
plot.yaxis.axis_label = "CTM Value"
show(plot)
```

```
# Putting it all together
def normalize(arr):
  return np.array(arr)/np.linalg.norm(arr)

plot = figure(**plot_options)
plot.line(range(1,len(sequences)+1), normalize(min_b_ent), color='red', legend_label='Minimum Entropy')
plot.line(range(1,len(sequences)+1), normalize(gzip_length), color='green', legend_label='gzip Compression')
plot.line(range(1,len(sequences)+1), normalize(bdm_vals), color='blue', legend_label='CTM Value')

plot.legend.location = "bottom_right"

plot.title.text = "Compression Length Analysis"
plot.xaxis.axis_label = "String Index"
plot.yaxis.axis_label = "Compression Length"
show(plot)
```
